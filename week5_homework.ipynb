```python
%pylab inline
from sklearn import model_selection, neighbors, datasets
```

```python
data = datasets.load_digits()
X, y = data.data, data.target
model = neighbors.KNeighborsClassifier()
train_scores, test_scores = model_selection.validation_curve(model, X, y, 'n_neighbors', range(1,31),cv=5)
```

```python
fig, ax = plt.subplots()
ax.plot(range(1,31),[mean(xx) for xx in train_scores],label='training error')
ax.plot(range(1,31),[mean(xx) for xx in test_scores],label='training error')
ax.legend()
```
