```python
%pylab inline
from sklearn import datasets, linear_model, neighbors, model_selection, neural_network,tree,ensemble
import warnings
from sklearn.svm import SVC
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')
```

```python
def alg_compare(X,y,title):
    model_list = [
        (linear_model.LogisticRegression(),'logiReg'),
        (SVC(kernel='linear'),'svm_lin'),
        (SVC(kernel='rbf'), 'svm_rbf'),
        (tree.DecisionTreeClassifier(),'DT'),
        (ensemble.RandomForestClassifier(),'RanFor'),
        (ensemble.AdaBoostClassifier(),'AdaBoo'),
        (ensemble.GradientBoostingClassifier(),'GDBT'),
        (neighbors.KNeighborsClassifier(n_neighbors=5),'knn5'),
        (neural_network.MLPClassifier(hidden_layer_sizes=(200,)),'NN'),
    ]
    x_axis = []
    for i in range(len(model_list)):
        x5 = []
        for j in range(1,6):
            x5.append(i*20+j)
        x_axis.append(x5)
    y_axis = []
    label_list=[]
    for model, label in model_list:
        result = model_selection.cross_val_score(model, X, y, cv=5)
        print(f'{label} cross_val_score {result}')
        y_axis.append(list(result))
        label_list.append(label)
    fig, ax = plt.subplots()
    for index, ls in enumerate(x_axis):
        ax.scatter(ls, y_axis[index],label=label_list[index])
    plt.xticks([i+2 for i in range(0,161,20)], label_list,fontsize=10)
    fig.suptitle(title,fontsize=16)
```

```python
data = datasets.load_iris()
X,y=data.data, data.target
alg_compare(X,y,'iris')
```

```python
data=datasets.load_digits()
X, y=data.data, data.target
alg_compare(X,y,'digits')
```

```python
X, y = datasets.make_circles(n_samples=500,noise=0.2,factor=0.2)
alg_compare(X,y,'circles')
```

```python
X,y=datasets.make_blobs(n_samples=500, centers = 3, cluster_std=2)
alg_compare(X,y,'blobs')
```

```python
X,y=datasets.make_moons(n_samples=500, noise=0.1)
alg_compare(X,y,'moons')
```
